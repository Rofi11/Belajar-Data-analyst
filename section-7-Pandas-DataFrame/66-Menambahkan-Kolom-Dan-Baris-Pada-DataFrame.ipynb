### 4. Menambahkan Kolom dan baris Pada DataFrame

```python
# Data awal
import pandas as pd
import numpy as np

# membuat data dictionary of list /object of array
data_kendaraan = {
    'Merek' : ['Toyota', 'Honda', 'Suzuki', 'Yamaha', 'Mitsubishi', 'Kawasaki', 'Daihatsu', 'Isuzu'],
    'Model' : ['Avanza', 'Civic', 'Ertiga', 'NMAX', 'Pajero', 'Ninja', 'Xenia', 'Panther'],
    'Tahun Produksi' : [2021, 2020, 2022, 2021, 2019, 2018, 2023, 2017],
    'Tipe' : ['Mobil', 'Mobil', 'Mobil', 'Motor', 'Mobil', 'Motor', 'Mobil', 'Mobil' ]
}

mob_index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

df_kendaraan = pd.DataFrame(data_kendaraan, index=mob_index)
df_kendaraan
```

#### Menambahkan Kolom

##### Menambahkan Kolom Baru

```python
# Siapkan List utk kolom baru
warna = ['Merah', 'Biru', 'Silver', 'Hitam', 'Hitam', 'Merah', 'Putih', 'Putih']
# masukan list dan buat key kolom
df_kendaraan['Warna Kendaraan'] = warna
df_kendaraan
```

##### Reassingment kolom atau Menggunakan data yang sudah ada
  * Reassingment -> kita bisa menimpa baris dengan data yang baru (Hati - hati dengan anam kolom tidak boleh sama)
  * Kita bisa menggabungka data kolom yang ada menjadi kolom baru

```python
df_kendaraan['Model dan Warna'] = df_kendaraan['Model'] + ' ' + df_kendaraan['Warna Kendaraan']
df_kendaraan
```

#### Menambahkan Baris/Index

##### Menambahkan Baris/Index Baru

```python
df_kendaraan.loc['I'] = ['Suzuki', 'Grand Vitara', 2022, 'Mobil', 'Biru', 'Grand Vitara Biru']
df_kendaraan
```
