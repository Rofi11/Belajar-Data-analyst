### 5. Menghapus Kolom dan Baris

```python
# Data awal
import pandas as pd
import numpy as np

# membuat data dictionary of list /object of array
data_kendaraan = {
    'Merek' : ['Toyota', 'Honda', 'Suzuki', 'Yamaha', 'Mitsubishi', 'Kawasaki', 'Daihatsu', 'Isuzu'],
    'Model' : ['Avanza', 'Civic', 'Ertiga', 'NMAX', 'Pajero', 'Ninja', 'Xenia', 'Panther'],
    'Tahun Produksi' : [2021, 2020, 2022, 2021, 2019, 2018, 2023, 2017],
    'Tipe' : ['Mobil', 'Mobil', 'Mobil', 'Motor', 'Mobil', 'Motor', 'Mobil', 'Mobil' ]
}

mob_index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

df_kendaraan = pd.DataFrame(data_kendaraan, index=mob_index)

# menambahkan kolom
df_kendaraan['Warna'] = ['Merah', 'Biru', 'Silver', 'Hitam', 'Hitam', 'Merah', 'Putih', 'Putih']
df_kendaraan['Model Warna'] = df_kendaraan['Model'] + ' ' + df_kendaraan['Warna']

df_kendaraan
```

#### Menghapus Kolom
  * dengan method -> .drop
  * disini ada beberapa parameter seperti column dan index, tinggal cek aja dengan nama_dataFrame.drop?
  * paramater column axis start dari 1 (nilai default)
  * parameter inplace=true, untuk mengesave perubahan Data
  * menghapus beberapa column dimasukan kedalam list []

```python
#  menghapus 1 kolom
df_kendaraan.drop(columns='Model Warna', axis=1, inplace=True)
df_kendaraan
```

```python
# Menghapus beberapa column
df_kendaraan.drop(columns=['Tipe', 'Warna'], axis=1, inplace=True)
df_kendaraan
```

#### Menghapus index/Baris
  * menggunakan method --> .drop
  * parameter nya index=
  * nilai axis nya tidak wajib, karena nilai default nya 0
  * menghapus multiple index dimasukan kedalam list []
  * parameter inplace=True --> untuk mensave data perubahan

```python
# Menghapus 1 baris/index
df_kendaraan.drop(index="A",inplace=True)
df_kendaraan
```

```python
# Menghapus beberapa index/baris
df_kendaraan.drop(index=["B", "C"])
```

#### Menghapus Kolom dan Index bersamaan
  * karena method drop secara default na punya parameter column dan index

```python
# Menghapus index dan columns
df_kendaraan.drop(columns=['Model'], index=['G'])
```
