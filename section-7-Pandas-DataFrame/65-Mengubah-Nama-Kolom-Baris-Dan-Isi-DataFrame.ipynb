### 3. Mengubah Nama kolom, Index, dan Isi DataFrame

```python
# Data awal
import pandas as pd
import numpy as np

# membuat data dictionary of list /object of array
data_kendaraan = {
    'Merek' : ['Toyota', 'Honda', 'Suzuki', 'Yamaha', 'Mitsubishi', 'Kawasaki', 'Daihatsu', 'Isuzu'],
    'Model' : ['Avanza', 'Civic', 'Ertiga', 'NMAX', 'Pajero', 'Ninja', 'Xenia', 'Panther'],
    'Tahun Produksi' : [2021, 2020, 2022, 2021, 2019, 2018, 2023, 2017],
    'Tipe' : ['Mobil', 'Mobil', 'Mobil', 'Motor', 'Mobil', 'Motor', 'Mobil', 'Mobil' ]
}

mob_index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

df_kendaraan = pd.DataFrame(data_kendaraan, index=mob_index)
df_kendaraan
```

#### Mengubah Nama kolom
  1. bisa dengan method .rename tapi tidak mereplace
    * masukan parameter columns -> buat dalam dictionary key-lama : key-baru
    * parameter inplace --> unutk menyimpan data/mereplace data

```python
# Cara mengubah Kolom
  # menggunakan parameter columns --> columns = {nama-kolom-sebelumnya : nama-kolom-yang-baru}
  # inplace --> berguna sebagai perintah untuk menyimpan data perubahan/ replace data kolom
df_kendaraan.rename(columns={
    'Merek':'Merk',
    'Tahun Produksi':'Tahun'
    },
    inplace=True
  )
df_kendaraan
```

#### Mengubah Baris/Index
  1. dengan method .rename
    * masukan parameter index utk mengubah baris --> buat dalam dictionary key-lama : key-baru
    * parameter inplace --> digunakan utk menyimpan perubahan / mereplace data
  2. dengan method set_index --> mereset index/merubah secara keseluruhan
    * Pertama : buat index baru
    * kedua : buat list yg dimasukan kedalam pd.Index (ini dibuat menjadi object index)
    * ketiga : baru kita bisa menggunakan set_index untuk mereset index
    * parameter inplace=True --> untuk menyimpan data perubahan

```python
# Cara 1 dengan method .rename
df_kendaraan.rename(index={
    'A' : 'Z',
    'B' : 'Y'
    },
    inplace=True
)
df_kendaraan
```

```python
# Cara 2 dengan set_index
# 1. membuat index dan masukan kedalam pd.Index utk membuat object of index
index_baru = pd.Index([1,2,3,4,5,6,7,8])
# 2. masukan kedalam set_index untuk merest seluruh index
  #  inplace jangan lupa, utk menyimpan perubahan
df_kendaraan.set_index(index_baru, inplace=True)
df_kendaraan
```

#### Mengubah Isi DataFrame
  1. Indexing dulu untuk mengarahkan data yang ingin di ubah
  2. menggunakan method loc, isi baris dan kolom yang mau dirubah
  3. ubah isi dengan data baru (data baru akan otomatis tersimpan)

```python
# 1. dari data dengan indexing
  # begini bisa dari hasil pencarian DataFrame
test1 = df_kendaraan.loc[[2], ['Tahun']] = 2025
print(test1)
  # begini juga bisa yang biasa
test2 = df_kendaraan.loc[3, 'Tahun'] = 2025
print(test2)

test3 = df_kendaraan.loc[3, 'Model'] = 'Avanza'
print(test3)

df_kendaraan
```
