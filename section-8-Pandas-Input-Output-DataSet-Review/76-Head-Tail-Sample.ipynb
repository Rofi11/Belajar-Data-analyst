#### Menyiapkan Data
* metode untuk mengambil sampel baris ada 3:
  1. Head -> default nya menampilkan 5 baris pertama
  2. Tail -> default nya menampilkan 5 baris akhir
  3. Sample -> default nya 1 data random
* Karena data biasa nya memeliki puluhan ribu baris, dan ini berguna untuk mengambil beberapa data saja yang dibutuhkan

```python
import pandas as pd

# copy path nya utk mencari data yang mau di baca
path = r"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\MISSION+5+FILES\provinsi_indonesia.csv"
df = pd.read_csv(path, sep=',')

# read data
df
```

##### Method Head
  Digunakan untuk menampilkan n baris pertama dari DataFrame
  * Secara default nya head akan mengambil 5 baris pertama saja
  * parameter bisa di isi dengan jumlah data yang pingin di tampilkan

```python
df.head(2)
```

##### Method Tail
  Digunakan untuk menampilkan n baris terakhir dari DataFrame
  * Secara default nya tail akan mengambil 5 baris terakhir saja
  * parameter bisa di isi dengan jumlah data yang pingin di tampilkan

```python
df.tail()
```

##### Method Sample
  Digunakan untuk menampilkan n baris secara acak dari DataFrame
  * default nya akan mengambil 1 data secara acak
  * parameter bisa di isi dengan jumlah data yang pingin di tampilkan (akan tetap random data yang di ambil nya)

```python
df.sample()
```
