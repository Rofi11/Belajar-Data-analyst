#### INPUT : Membaca Data

```python
import pandas as pd
```

##### 1. Membaca Data dari File CSV (Comma Separated Value)
  * Untuk membaca/mencari data harus path spesifik dan copy path nya dengan klik icon tiga di samping file, aggar rapih masukan ke dalam variabel
  * parameter ->  
    1. sep= adalah parameter pemisah data nya, harus sama yg dimasukan nya dengan yg di file (cek dulu jangan lupa, format nya wajib string juga)
    2. index_col= adalah parameter untuk membuat column menjadi index baris("harus string tipe data nya juga)
    3. dalam VS CODE cara baca rada unik tambahkan r di depan string agar bisa di baca

```python
# copy path nya utk mencari data yang mau di baca
path = r"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\MISSION+5+FILES\provinsi_indonesia.csv"
df1 = pd.read_csv(path, sep=',')

# read data
df1
```

##### 2. Membaca Data dari File Excel
  * Parameter
    1. sheet_name= adalah parameter untuk membaca sheet mana yang ada di excel yang akan di baca oleh pandas (tipe data harus string juga)

```python
path2 = r"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\MISSION+5+FILES\provinsi_indonesia.xlsx"
df2 = pd.read_excel(path2, sheet_name='provinsi')

# read data
df2
```

##### 3. Membaca Data dari File JSON


```python
path3 = r"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\MISSION+5+FILES\provinsi_indonesia.json"
df3 = pd.read_json(path3)
# read data
df3
```

#### OUTPUT : Menulis Data

##### 1. Menulis kedalam CSV
  * menggunakan .to_csv() -> contoh data.to_csv(path)
    1. buat nama file dengan format csv
    2. path folder (tempat menyimpan data)
    3. masukan dengan .to_csv()

```python
# 1. buat nama file csv
nama_csv = 'provinsi_csv_baru.csv'
# 2. path folder (tempat menyimpan data) dan masukan nama file nya
path_folder = rf"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\folder_menyimpan_data_baru/{nama_csv}"
# 3. jalankan code untuk memasukan data
df1.to_csv(path_folder)
```

##### 2. Menulis Kedalam excel
  * menggunakan .to_excel()-> contoh data.to_excel(path)
    1. buat nama file dengan format xlsx
    2. path folder (tempat menyimpan data)
    3. masukan dengan .to_excel()

```python
# 1. buat nama file excel
nama_excel = 'provinsi_excel_baru.xlsx'
# 2. path folder (tempat menyimpan data) dan masukan nama file nya
path_folder = rf"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\folder_menyimpan_data_baru/{nama_excel}"
# 3. jalankan code untuk memasukan data
df2.to_excel(path_folder)
```

##### 3. Menulis kedalam Json
 * menggunakan to_json() -> contoh data.to_json(path)
    1. buat nama file dengan format json
    2. path folder (tempat menyimpan data)
    3. masukan dengan .to_json()

```python
# 1. buat nama file json
nama_json = 'provinsi_json_baru.json'
# 2. path folder (tempat menyimpan data) dan masukan nama file nya
path_folder = rf"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\folder_menyimpan_data_baru/{nama_json}"
# 3. jalankan code untuk memasukan data
df3.to_json(path_folder)
```
