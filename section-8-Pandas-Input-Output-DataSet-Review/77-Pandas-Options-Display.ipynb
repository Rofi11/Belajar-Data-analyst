#### **Pandas.Options.Display**
pandas.options.display adalah salah satu atribut dalam pustaka pandas yang digunakan untuk mengatur berbagai opsi tampilan DataFrame dan objek lainya

```python
import pandas as pd

path_pokemon= r"C:\Rofi belajar\Udemy Data Analyst\section-8-Pandas-Input-Output-DataSet-Review\MISSION+5+FILES\Pokemon.csv"

df_pokemon = pd.read_csv(path_pokemon, index_col='Num')
df_pokemon
```

##### **display.max_rows**
Opsi ini mengatur jumlah baris maksimum yang akan ditampilkan saat sebuah DataFrame dicetak. Misalnya, jika diatur ke 10 hanya 10 baris pertama yang akan di tampilkan.
  * Jika max_rows = None, maka seluruh baris akn di tampilkan tanpa ada yang di singkat/hiding

```python
pd.options.display.max_rows = 4
df_pokemon
```

##### **display.max_columns**
Opsi ini mengatur jumlah kolom maksimum yang akan di tampilkan. Misalnya, jika diatur ke 5, hanya 5 kolom pertama yang akan ditampilkan.
  * Jika max_columns = None, maka akan menampilkan seluruh data tanpa ada yang di singkat

```python
pd.options.display.max_columns = None
df_pokemon
```

##### **display.width**
Opsi ini mengatur lebar maksimum dalam karakter dari tampilan DataFrame. Misalnya, jika diatur ke 80, tampilan DataFrame tidak akan melebihi 80 karakter dalam satu baris
  * disin pun bisa di isi oleh **None**
  * fokus width mengatur jumlah karakter yg bisa di tampilan di setiap column dan baris

```python
pd.options.display.max_colwidth = 3
df_pokemon
```

##### **diplay.precision**
Opsi ini mengatur jumlah angka desimal yang akan ditampilkan untuk angka floating point.

```python
pd.options.display.max_rows = None

# untuk mengatur jumlah angka di belakang koma
pd.options.display.precision = 2
df_pokemon.mean(numeric_only=True)
```
